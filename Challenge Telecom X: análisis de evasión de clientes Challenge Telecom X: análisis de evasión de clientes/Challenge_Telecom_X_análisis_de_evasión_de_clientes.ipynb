# **Importando librerias**

```python
import pandas as pd # Importa la librería pandas para manipulación y análisis de datos.
```

# **Extracción del Archivo Tratado**

```python
df=pd.read_csv('/content/df_normalizado_prueba.csv') # Carga el archivo CSV en un DataFrame de pandas.
df # Muestra las primeras y últimas filas del DataFrame para una vista previa.
```

# **Eliminación de Columnas Irrelevantes**

```python
df.drop('customerID', axis=1, inplace=True) # Elimina la columna 'customerID' porque es un identificador único y no es útil para el modelo predictivo.
df # Muestra el DataFrame modificado.
```

```python
df.info() # Muestra un resumen del DataFrame, incluyendo el número de entradas no nulas y los tipos de datos de cada columna.
```

# **Haciendo el encoding**

```python
df_encoded = pd.get_dummies(df, drop_first=True) # Aplica One-Hot Encoding a las columnas categóricas para convertirlas en formato numérico.
# `drop_first=True` evita la multicolinealidad eliminando una de las categorías.
df_encoded.head() # Muestra las primeras filas del DataFrame codificado.
```

```python
df_encoded.info() # Muestra un resumen del DataFrame codificado, verificando los nuevos tipos de datos (principalmente booleanos).
```

# **Verificación de la Proporción de Cancelación (Churn)**

```python
df['Churn'].value_counts(normalize=True).round(2) # Calcula la proporción de cada clase en la columna 'Churn' (Sí/No) para verificar el desbalance de clases.
```

# **Balanceo de Clases**

```python
x = df_encoded.drop('Churn_Yes', axis=1) # Separa las características (variables independientes) del DataFrame.
y = df_encoded['Churn_Yes'] # Asigna la columna 'Churn_Yes' como la variable objetivo (dependiente).
```

```python
from imblearn.over_sampling import SMOTE # Importa la librería SMOTE para el balanceo de clases.

sm = SMOTE(random_state=42) # Inicializa el algoritmo SMOTE con una semilla para reproducibilidad.
x_res, y_res = sm.fit_resample(x, y) # Aplica SMOTE para sobremuestrear la clase minoritaria y balancear el dataset.
```

```python
y_res.value_counts() # Muestra el conteo de cada clase en la variable objetivo después de aplicar SMOTE, verificando que estén balanceadas.
```

# **Análisis de Correlación**

```python
corr = df_encoded.corr() # Calcula la matriz de correlación de todas las variables en el DataFrame codificado.
corr # Muestra la matriz de correlación completa.
```

```python
corr_churn = corr['Churn_Yes'].sort_values(ascending=False) # Extrae la columna de correlación con 'Churn_Yes' y la ordena de mayor a menor.
corr_churn # Muestra las correlaciones con la variable objetivo.
```

```python
import seaborn as sns # Importa seaborn para visualización de datos.
import matplotlib.pyplot as plt # Importa matplotlib para la creación de gráficos.

plt.figure(figsize=(8,12)) # Define el tamaño de la figura para el gráfico.
corr_churn.drop('Churn_Yes').plot(kind='barh') # Crea un gráfico de barras horizontales de las correlaciones, excluyendo la correlación de Churn_Yes consigo mismo.
plt.title('Correlación con Churn') # Establece el título del gráfico.
plt.show() # Muestra el gráfico.
```

# **Análisis Dirigido**

## **Boxplot: Tiempo de contrato por Cancelación**

```python
plt.figure(figsize=(8,5)) # Define el tamaño de la figura.
sns.boxplot(x='Churn_Yes', y='tenure', data=df_encoded) # Crea un boxplot para visualizar la distribución del 'tenure' (tiempo de contrato) en relación con 'Churn_Yes'.
plt.title('Tiempo de contrato (tenure) vs Cancelación (Churn)') # Establece el título del gráfico.
plt.xlabel('Canceló (1 = Sí, 0 = No)') # Etiqueta el eje X.
plt.ylabel('Meses de contrato') # Etiqueta el eje Y.
plt.show() # Muestra el gráfico.
```

## **Boxplot: Gasto Total vs Cancelación**

```python
plt.figure(figsize=(8,5)) # Define el tamaño de la figura.
sns.boxplot(x='Churn_Yes', y='Charges.Total', data=df_encoded) # Crea un boxplot para visualizar la distribución de 'Charges.Total' (gasto total) en relación con 'Churn_Yes'.

plt.title('Gasto Total vs Cancelación') # Establece el título del gráfico.
plt.xlabel('Canceló (1 = Sí, 0 = No)') # Etiqueta el eje X.
plt.ylabel('Gasto Total') # Etiqueta el eje Y.
plt.show() # Muestra el gráfico.
```

# **Separación de Datos**

```python
from sklearn.model_selection import train_test_split # Importa la función para dividir los datos en conjuntos de entrenamiento y prueba.
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y) # Divide los datos en conjuntos de entrenamiento (80%) y prueba (20%).
# `random_state` asegura reproducibilidad y `stratify=y` mantiene la proporción de clases en los conjuntos.
```

# **Creación de Modelos**

## **Modelo 1: Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier # Importa la clase DecisionTreeClassifier para crear un modelo de Árbol de Decisión.
```

```python
dt_model = DecisionTreeClassifier(random_state=42) # Inicializa el modelo de Árbol de Decisión con una semilla para reproducibilidad.
dt_model.fit(X_train, y_train) # Entrena el modelo utilizando los datos de entrenamiento.
```

```python
y_pred_dt = dt_model.predict(X_test) # Realiza predicciones sobre el conjunto de prueba utilizando el modelo de Árbol de Decisión.
```

## **Modelo 2: Random Forest**



```python
from sklearn.ensemble import RandomForestClassifier # Importa la clase RandomForestClassifier para crear un modelo de Bosque Aleatorio.
```

```python
rf_model = RandomForestClassifier(random_state=42) # Inicializa el modelo de Bosque Aleatorio con una semilla para reproducibilidad.
rf_model.fit(X_train, y_train) # Entrena el modelo utilizando los datos de entrenamiento.
```

```python
y_pred_rf = rf_model.predict(X_test) # Realiza predicciones sobre el conjunto de prueba utilizando el modelo de Bosque Aleatorio.
```

# **Evaluacion Modelos**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report # Importa las métricas de evaluación de modelos.
```

## **Decision Tree**

```python
print("=== DECISION TREE ===") # Imprime un encabezado para los resultados del modelo.

print("Accuracy:", accuracy_score(y_test, y_pred_dt)) # Calcula y muestra la exactitud del modelo.
print("Precision:", precision_score(y_test, y_pred_dt)) # Calcula y muestra la precisión del modelo.
print("Recall:", recall_score(y_test, y_pred_dt)) # Calcula y muestra el recall (sensibilidad) del modelo.
print("F1 Score:", f1_score(y_test, y_pred_dt)) # Calcula y muestra el F1-Score del modelo.

print("\nMatriz de Confusión:") # Imprime un encabezado.
print(confusion_matrix(y_test, y_pred_dt)) # Calcula y muestra la matriz de confusión.

print("\nReporte completo:") # Imprime un encabezado.
print(classification_report(y_test, y_pred_dt)) # Muestra un reporte completo de clasificación con todas las métricas.
```

## **Random Forest**

```python
print("=== RANDOM FOREST ===") # Imprime un encabezado para los resultados del modelo.

print("Accuracy:", accuracy_score(y_test, y_pred_rf)) # Calcula y muestra la exactitud del modelo.
print("Precision:", precision_score(y_test, y_pred_rf)) # Calcula y muestra la precisión del modelo.
print("Recall:", recall_score(y_test, y_pred_rf)) # Calcula y muestra el recall (sensibilidad) del modelo.
print("F1 Score:", f1_score(y_test, y_pred_rf)) # Calcula y muestra el F1-Score del modelo.

print("\nMatriz de Confusión:") # Imprime un encabezado.
print(confusion_matrix(y_test, y_pred_rf)) # Calcula y muestra la matriz de confusión.

print("\nReporte completo:") # Imprime un encabezado.
print(classification_report(y_test, y_pred_rf)) # Muestra un reporte completo de clasificación con todas las métricas.
```

## **Decision Tree**

```python
importances_dt = dt_model.feature_importances_ # Obtiene la importancia de las características del modelo Decision Tree.
features = x.columns # Obtiene los nombres de las columnas (características).

df_importance_dt = pd.DataFrame({
    'Variable': features,
    'Importancia': importances_dt
}).sort_values(by='Importancia', ascending=False) # Crea un DataFrame con las variables y sus importancias, ordenado de mayor a menor.

df_importance_dt.head(10) # Muestra las 10 variables más importantes.
```

```python
import matplotlib.pyplot as plt # Importa matplotlib para la creación de gráficos.

plt.figure(figsize=(8,6)) # Define el tamaño de la figura.
plt.barh(df_importance_dt['Variable'][:10],
         df_importance_dt['Importancia'][:10]) # Crea un gráfico de barras horizontales de las 10 variables más importantes.

plt.gca().invert_yaxis() # Invierte el eje Y para que la variable más importante quede arriba.
plt.title('Top 10 Variables Más Importantes - Decision Tree') # Establece el título del gráfico.
plt.xlabel('Importancia') # Etiqueta el eje X.
plt.ylabel('Variable') # Etiqueta el eje Y.
plt.show() # Muestra el gráfico.
```

# **Informe de Análisis Predictivo de Cancelación de Clientes (Churn)**

## 1. Preparación y comprensión de los datos

Se trabajó con un conjunto de datos compuesto por 7043 registros de clientes y 21 variables relacionadas con características demográficas, servicios contratados, tipo de contrato y cargos económicos. Inicialmente se realizó una revisión de calidad de datos, verificando que no existieran valores nulos y que los tipos de datos fueran consistentes.

Posteriormente se eliminaron columnas que no aportaban valor predictivo, como el identificador único del cliente (`customerID`), ya que no contiene información relevante para la predicción. Las variables categóricas fueron transformadas a formato numérico mediante **One-Hot Encoding**, permitiendo que los algoritmos de machine learning pudieran procesarlas correctamente.

Se analizó la proporción de la variable objetivo (cancelación o *Churn*), observándose un desbalance moderado, donde la mayoría de clientes permanecen activos y una menor proporción cancela el servicio. Este aspecto es importante porque puede influir en la interpretación de métricas como la exactitud.



### 2. Análisis exploratorio de datos

Se realizó un análisis de correlaciones y visualizaciones gráficas para identificar relaciones entre variables numéricas y la cancelación. Los boxplots mostraron que:

* Los clientes con menor tiempo de permanencia (tenure) tienden a cancelar con mayor frecuencia.
* El gasto total acumulado suele ser menor en clientes que cancelan, lo que sugiere que los clientes nuevos o con menor compromiso económico presentan mayor riesgo de abandono.

Estas observaciones permitieron identificar patrones de comportamiento relevantes antes del modelado.


### 3. División del conjunto de datos

El dataset fue dividido en 80 % para entrenamiento y 20 % para prueba, utilizando muestreo estratificado para mantener la proporción de clientes que cancelan y los que permanecen. Esta división permitió evaluar el rendimiento de los modelos en datos no vistos previamente.


### 4. Modelos predictivos implementados

Se seleccionaron dos modelos que no requieren normalización de datos, debido a que se basan en reglas de decisión y no en cálculos de distancia:

* Decision Tree (Árbol de Decisión)
* Random Forest

La elección se fundamenta en que estos modelos son robustos ante diferencias de escala entre variables y son adecuados para datasets tabulares como el de churn.


### 5. Evaluación de modelos

Los modelos fueron evaluados utilizando las métricas:

* **Exactitud (Accuracy)**
* **Precisión**
* **Recall**
* **F1-Score**
* **Matriz de Confusión**

Se observó que Random Forest presentó un mejor equilibrio entre precisión y recall, logrando una mayor capacidad de generalización. El Decision Tree mostró un rendimiento aceptable, pero con mayor riesgo de overfitting, debido a su tendencia a ajustarse en exceso a los datos de entrenamiento.

En problemas de churn, el **Recall** y el **F1-Score** son especialmente importantes, ya que interesa identificar correctamente a los clientes que podrían cancelar, más que simplemente maximizar la exactitud global.

### 6. Importancia de variables

El análisis de importancia de variables, especialmente mediante Random Forest, permitió identificar los factores más influyentes en la cancelación:

* **Tipo de contrato mensual (Month-to-Month):**
  Los clientes sin contratos a largo plazo presentan mayor probabilidad de cancelación.
* **Tiempo de permanencia (Tenure):**
  A mayor antigüedad, menor probabilidad de abandono.
* **Cargos mensuales y totales:**
  Reflejan el nivel de compromiso económico del cliente.
* **Tipo de servicio de internet (Fiber Optic en algunos casos):**
  Puede asociarse a expectativas de servicio más altas.

Estas variables mostraron mayor contribución en la reducción de impureza dentro de los árboles de decisión, lo que indica su relevancia en la clasificación.

### 7. Factores principales que influyen en la cancelación

A partir del análisis realizado, los factores más determinantes son:

1. **Estabilidad contractual** (tipo de contrato).
2. **Antigüedad del cliente** (tenure).
3. **Compromiso económico** (gastos mensuales y totales).
4. **Características del servicio contratado**.

Estos factores indican que la cancelación está fuertemente relacionada con la falta de compromiso a largo plazo y con la etapa inicial del ciclo de vida del cliente.


### 8. Estrategias de retención propuestas

Con base en los hallazgos, se pueden proponer las siguientes estrategias:

* Incentivar contratos a largo plazo mediante descuentos o beneficios exclusivos.
* Programas de fidelización temprana dirigidos a clientes nuevos.
* Ofertas personalizadas según gasto mensual.
* Mejoras en la calidad del servicio para clientes con planes premium.
* Seguimiento proactivo de clientes con bajo tenure y contratos mensuales.


### 9. Conclusión General

El presente análisis permitió no solo construir modelos predictivos capaces de identificar clientes con alta probabilidad de cancelación, sino también comprender los factores estructurales que explican dicho comportamiento. La combinación de análisis exploratorio, modelado supervisado y estudio de importancia de variables proporcionó una visión integral del fenómeno de churn.

Desde el punto de vista del rendimiento predictivo, **Random Forest** demostró ser el modelo más robusto y estable. Su capacidad para reducir la varianza mediante el ensamble de múltiples árboles le permitió obtener mejores métricas de F1-Score y Recall en comparación con el Decision Tree individual. Esto es particularmente relevante en problemas de cancelación, donde detectar correctamente a los clientes en riesgo es más importante que maximizar únicamente la exactitud global.

El Decision Tree, aunque interpretable y útil para comprender reglas de decisión, mostró mayor sensibilidad al sobreajuste (overfitting), lo cual es coherente con su naturaleza determinística y su tendencia a adaptarse excesivamente a los datos de entrenamiento cuando no se regula adecuadamente la profundidad o los criterios de división.

En cuanto a los factores explicativos, el análisis confirmó que la cancelación está principalmente asociada a:

* Bajo tiempo de permanencia (clientes recientes).
* Contratos mensuales sin compromiso a largo plazo.
* Menor acumulación de gasto total.
* Determinados tipos de servicios contratados.

Estos resultados sugieren que la cancelación no es un fenómeno aleatorio, sino que está fuertemente vinculada al ciclo de vida del cliente y al nivel de compromiso contractual. En términos estratégicos, esto implica que la empresa puede intervenir tempranamente en segmentos específicos para reducir el abandono.
